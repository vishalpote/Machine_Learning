```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Set1')
plt.colorbar()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
